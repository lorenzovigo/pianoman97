```python
import numpy as np
import math
```

```python
def generate_points(angle=45):
    m = [0., 0.]
    angle = 45 * math.pi / 180
    rot = np.array([[math.cos(angle), -math.sin(angle)], [math.sin(angle), math.cos(angle)]])
    lamb = np.array([[100,0],[0,1]])
    s = np.matmul(rot, np.matmul(lamb, rot.transpose()))
    c = np.random.multivariate_normal(m,s,100)
    return c
```

Let's define the function, its gradient and the gradient descent algorithm

```python
def Q(w0, w1):
    return 1/2 * np.sum((w0 * c[:,0] + w1 - c[:,1])**2)

def grad_Q(w0, w1):
    return np.array([np.sum((w0 * c[:,0] + w1 - c[:,1]) * c[:,0]), np.sum(w0 * c[:,0] + w1 - c[:,1])])

def gradient_descent_adapt_alpha(f, grad_f, w0, f_tol=1e-3, grad_tol=1e-5):
    x = [w0]
    # Iterating until one of the stop criteria is fulfilled
    while True:
        alpha = 1
        # Compute new alphas until the needed condition is true.
        while f(*(w0 - alpha * grad_f(w0[0],w0[1]))) >= f(*w0):
            alpha /= 2

        # Same formula to compute the next iteration point
        w0 = w0 - alpha * grad_f(w0[0], w0[1])
        x.append(w0)
        
        # If one of the stopping criterion is satisfied, we return the history of points
        if np.abs(f(*x[-1]) - f(*x[-2])) < f_tol or np.linalg.norm(grad_f(*x[-1])) < grad_tol:
            return np.array(x)
            
    return np.array(x)
```

Here we will generate a set of points, apply the gradient descent to get the optimal parameters and plot the points and the line regressor

```python
c = generate_points()

w0 = np.random.rand(2)

path = gradient_descent_adapt_alpha(Q, grad_Q, w0)
print("The optimal parameters we obtained are:", path[-1])

import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-20,20,40)
yhat = path[-1][0]*x + path[-1][1]

plt.scatter(c[:,0],c[:,1])
plt.plot(x, yhat, 'r')
```

Again, we will generate a set of points, change the second one with $[-40,20]$ (an outlier), apply the gradient descent to get the optimal parameters and plot the points and the line regressor

```python
c = generate_points()
c[1] = [-40, 20]

w0 = np.random.rand(2)

path = gradient_descent_adapt_alpha(Q, grad_Q, w0)
print("The optimal parameters we obtained are:", path[-1])

import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-40,40,80)
yhat = path[-1][0]*x + path[-1][1]

plt.scatter(c[:,0],c[:,1])
plt.plot(x, yhat, 'r')
```

Comparing with the previous plot, we could see that the outlier we've inserted is degenerating our linear regressor

Now, let's modify some points (to be outliers), 10, to see the influence of the outliers

```python
c = generate_points()

for i in (np.random.rand(10)*100).astype(int):
    c[i] = (np.random.rand(2) - 0.5) * 80

w0 = np.array([0.,0.])

path = gradient_descent_adapt_alpha(Q, grad_Q, w0)
print("The optimal parameters we obtained are:", path[-1])

import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-40,40,80)
yhat = path[-1][0]*x + path[-1][1]

plt.scatter(c[:,0],c[:,1])
plt.plot(x, yhat, 'r')
```

We've changed only 10 points to be outliers and the linear regressor is not resistent to them, the result varies is a lot.

Let's define the LSE and Cauchy fucntions

```python
def plot_lse(u):
    return 1/2 * u**2

def plot_cauchy(u, k = 1):
    return k**2 / 2 * np.log(1 + (u/k)**2)
```

And plot the on the same graphic

```python
x = np.linspace(-10,10,80)

plt.plot(x, plot_lse(x), 'r')
plt.plot(x, plot_cauchy(x), 'b')
```

Let's define the cauchy function and its gradient

```python
# Set globally for experiment 2.5
k = 1
def cauchy(w0, w1):
    return np.sum(k**2 / 2 * np.log(1 + ((w0 * c[:,0] + w1 - c[:,1]) / k)**2))

def grad_cauchy(w0, w1):
    aux = w0 * c[:,0] + w1 - c[:,1]
    return np.array([np.sum((c[:,0] * aux) / (1 + (aux / k)**2)), np.sum(aux / (1 + (aux / k)**2))])
```

Let's generate a set of points, apply the gradient descent using cauchy and plot the linear regressor

```python
c = generate_points()

rand_w = np.random.rand(2)

# Cauchy
path = gradient_descent_adapt_alpha(cauchy, grad_cauchy, rand_w)
print("The optimal parameters we obtained are:", path[-1])

x = np.linspace(-20,20,40)
yhat = path[-1][0]*x + path[-1][1]

plt.scatter(c[:,0],c[:,1])
plt.plot(x, yhat, 'r')
```
