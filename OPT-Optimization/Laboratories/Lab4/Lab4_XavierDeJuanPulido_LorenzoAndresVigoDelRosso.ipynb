We define the function $f$ in our problem and the equality constraint $h$. We will also compute their gradients and its hessian.

```python
from math import exp as e
import numpy as np

def f(x_1, x_2):
    return e(3*x_1) + e(-4*x_2)

def h(x_1, x_2):
    return x_1**2 + x_2**2 - 1

def grad_f(x_1, x_2):
    return np.array([3*e(3*x_1), -4*e(-4*x_2)])

def grad_h(x_1, x_2):
    return np.array([2*x_1, 2*x_2])

def hess_f(x_1, x_2):
    return np.array([[9*e(3*x_1), 0], [0, 16*e(-4*x_2)]])

def hess_h(x_1, x_2):
    return np.array([[2, 0], [0, 2]])
```

As stated in the laboratory, $\mathcal{L}(x, \lambda) = f(x) - \lambda h(x)$. We will define this function, its gradient and its hessian.

```python
# TODO we are considering that the gradient of L should only be respect to x and not to landa also

def lag(x_1, x_2, landa):
    return f(x_1, x_2) - landa*h(x_1, x_2)

def grad_x_lag(x_1, x_2, landa):
    return grad_f(x_1, x_2) - landa*grad_h(x_1, x_2)

def hess_x_lag(x_1, x_2, landa):
    return hess_f(x_1, x_2) - landa*hess_h(x_1, x_2)
```

Now, let's define the Newton based iterative method to solve the problem.

```python
def solveNewtonBased(x_1, x_2, landa, alpha=1, eps=1e-5, MAX_ITER=100, verbose=False):

    for i in range(MAX_ITER):
        #¬†Build matrix A
        gh = grad_h(x_1, x_2)[:,np.newaxis]
        A = np.concatenate((
            np.concatenate((hess_x_lag(x_1, x_2, landa), -gh), axis=1),
            np.concatenate((-gh.T, np.zeros((1,1))), axis=1),
        ))
        #¬†Build vector b
        b = np.concatenate((
            -grad_x_lag(x_1, x_2, landa),
            np.array([h(x_1, x_2)]),
        ))
        #¬†Solve the system
        delta = np.linalg.solve(A, b)

        #¬†Update variables
        x_1 = x_1 + alpha*delta[0]
        x_2 = x_2 + alpha*delta[1]
        landa = landa + alpha*delta[2]

        if np.linalg.norm(grad_x_lag(x_1, x_2, landa)) < eps:
            print('Break by Lagrangian gradient')
            break
    if verbose:
        print('Iterations: {}'.format(i))
        print('x = (x_1, x_2) = ({0:.5f}, {1:.5f}), landa = {2:.5f}'.format(x_1, x_2, landa))
    return x_1, x_2, landa

x_1, x_2, landa = -1, 1, -1
_ = solveNewtonBased(x_1, x_2, landa, verbose=True)
        
```

We can observe that the solution is reached in only 2 iterations and matches with the solution given in the statement.

Let's define the merit function and its gradient.

```python
def merit(x_1, x_2, ro=10):
    return f(x_1, x_2) + ro*h(x_1, x_2)**2

def grad_merit(x_1, x_2, ro=10):
    return grad_f(x_1, x_2) + 2 * ro * h(x_1, x_2) * grad_h(x_1, x_2)
```

We will use the gradient descent implemented in previous laboratories but with gradient normalization.

```python
def gradient_descent(f, grad_f, w0, f_tol=1e-3, grad_tol=1e-5):
    x = [w0]
    # Iterating until one of the stop criteria is fulfilled
    while True:
        gradient_of_f = grad_f(w0[0], w0[1])
        grad_normalized = gradient_of_f / np.linalg.norm(gradient_of_f)
        alpha = 1
        # Compute new alphas until the needed condition is true.
        while f(*(w0 - alpha * grad_normalized)) >= f(*w0):
            alpha /= 2

        # Same formula to compute the next iteration point
        w0 = w0 - alpha * grad_normalized
        x.append(w0)
        
        # If one of the stopping criterion is satisfied, we return the history of points
        gradient_of_f = grad_f(w0[0], w0[1])
        grad_normalized = gradient_of_f / np.linalg.norm(gradient_of_f)
        if np.abs(f(*x[-1]) - f(*x[-2])) < f_tol or np.linalg.norm(grad_normalized) < grad_tol:
            return np.array(x)
```

Let's test the previous method with a point far away from the solution to find a point that is closer to the optimal solution.

```python
#¬†We are taking the first far away point defined in the previous exercise
p_try = far_away_points[0]
sol_approx = gradient_descent(merit, grad_merit, p_try[:2])
print("After running gradient descent on Merit function we get the approximation:", sol_approx[-1])
```
